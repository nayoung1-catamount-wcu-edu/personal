```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np  

import pandas as pd
```

```python
# This will make the plots larger.
# Comment it out if you don't want that
plt.rcdefaults()
```

### Read Country data into dataframe and pivot it

```python
df = pd.read_csv('Data\\Country_Data.csv')
df.head()
```

```python
print("Number of Rows in df:  ", df.shape[0])
print("Number of Columns  in df:  ", df.shape[1])

# df['country'].unique()
```

```python
df['country'].unique()
```

### Pivot Data to *Wide* Shape

```python
# Pivot:  value = gdpPerCapita
df_wide_GDP = df.pivot(index='year', columns='country', values='gdpPerCapita')
df_wide_GDP.head(5)
```

# Single Line Charts

```python
plt.plot(df_wide_GDP.index,
         df_wide_GDP['Albania'])

plt.show()
```

# Multi-Line Charts

```python
plt.plot(df_wide_GDP.index, 
         df_wide_GDP['Albania'], 
         label='Albania')

plt.plot(df_wide_GDP.index, 
         df_wide_GDP['Australia'], 
         label='Australia')

plt.legend()
plt.title('Multi-Line Plot:  Albania and Australia')

plt.show()
```

# Line Formatting

### Line Styles
We can have solid, dashed, dotted or dash-dot lines  

Line formats: https://matplotlib.org/gallery/lines_bars_and_markers/line_styles_reference.html 

```python
plt.plot(df_wide_GDP.index, 
         df_wide_GDP['Australia'],  
         color='green',
         linestyle=':'
        )
plt.show()
```

```python
plt.plot(df_wide_GDP.index, 
         df_wide_GDP['Australia'],  
         color='green',
         linestyle='--'
        )

plt.show()
```

```python
plt.plot(df_wide_GDP.index, 
         df_wide_GDP['Australia'],  
         color='green',
         linestyle=':',
         linewidth=3
        )

plt.show()
```

### Markers 
The 'd' denotes small diamonds. For all the marker styles check out this page: <br />
https://matplotlib.org/1.4.1/api/markers_api.html

```python
plt.plot(df_wide_GDP.index, 
         df_wide_GDP['Australia'], 
         color='green',
         marker='d'
        )
plt.show()
```

```python
plt.plot(df_wide_GDP.index, 
         df_wide_GDP['Australia'], 
         marker='d',
         linestyle=':',
         color='red')

plt.title('GDP per Capita:  Australia')

plt.show()
```

### Axis Labels

```python
plt.plot(df_wide_GDP.index, 
         df_wide_GDP['Australia'])

plt.xlabel('x', fontsize=15, color='green')
plt.ylabel('2*x', fontsize=15, color='green')

plt.show()
```

### Ticks and Tick Labels

```python
# Modify the Y axis by:
# 1. Changing the color of the ticks  
# 2. Changing the color of the tick label  
# 3. Changing the size of the tick label
plt.tick_params(axis='x',
                color='red',
                labelcolor='blue',
                labelsize='xx-large')
```

#### We can also get rid of the ticks and tick labels
Here we modify the X axis:
* <b>bottom=false</b> gets rid of the ticks
* <b>labelbottom=false</b> removes the labels

```python
plt.tick_params(axis = 'x',
                bottom=False,
                labelbottom=False
               )
```

#### Plot with Labeled Axes

```python
plt.plot(df_wide_GDP.index, 
         df_wide_GDP['Australia']
        )

plt.xlabel('Year', fontsize=15, color='green')
plt.ylabel('GDP per Capita', fontsize=15, color='purple')

plt.show()
```

### Legends

```python
plt.plot(df_wide_GDP.index, 
         df_wide_GDP['Panama'], 
        label='GDP per Capita for Panama'
        )

plt.legend()
plt.show()
```

#  Colors
- The colors of each plot is chosen by iterating over a color palette. 
- The default palette color order is:  
  - blue  
  - orange  
  - green  
  - red  
  - purple  
  - brown  
  - pink  
  - gray  
  - olive  
  - cyan 
  

The colors and codes for Matplotlib are here:
https://matplotlib.org/2.0.2/api/colors_api.html

The full list of named colors is here:
https://matplotlib.org/examples/color/named_colors.html

```python
# We have specified the full name of the green color
# Magenta has been specified in shorthand ('g' is short for green) 

plt.plot(df_wide_GDP.index, 
         df_wide_GDP['Ecuador'], 
         label='Ecuador',
         color='green'
        )

plt.plot(df_wide_GDP.index, 
         df_wide_GDP['Colombia'], 
         label='Colombia',
         color='m'
        )

plt.legend()
plt.title('Multi-Line Plot:  Ecuador and Colombia')

plt.show()
```

# Bar Charts

### Vertical Bar Charts

```python
barWidth = 4

# plot as Bar chart
plt.bar(df_wide_GDP.index, 
        df_wide_GDP['Venezuela'],
        width=barWidth,
        align='center')

#plt.legend()
plt.title('Venezuela GDP per capita')

plt.show()
```

### Horizontal Bar Charts

```python
plt.barh(df_wide_GDP.index, 
         df_wide_GDP['Argentina'],
         height=4)

plt.title('Argentina')

plt.show()
```

# Grouped Bar Charts

```python
x_values = df_wide_GDP.index
y_values = df_wide_GDP['Cuba']
y_values2 = df_wide_GDP['Haiti']

y_values_label = 'Cuba'
y_values2_label = 'Haiti'

# set width of bar
barWidth = 0.25

# Set position of bar on X axis
r1 = np.arange(len(y_values))
r2 = [x + barWidth for x in r1]

# First country to plot
plt.bar(r1, 
        y_values, 
        color='#7f6d5f', 
        width=barWidth, 
        edgecolor='white', 
        label=y_values_label)


# Second country to plot
plt.bar(r2, 
        df_wide_GDP['Haiti'], 
        color='#557f2d', 
        width=barWidth, 
        edgecolor='white', 
        label=y_values2_label)

# Add xticks on the middle of the group bars
plt.xlabel('Years', fontweight='bold')
plt.xticks([r + barWidth for r in range(len(y_values))], x_values)

plt.legend()
plt.title(y_values_label + " and " + y_values2_label + " per capita GDP")

plt.show()
```

# Stacked Bar Charts

```python
# Heights of bars1 + bars2
bars = np.add(y_values, y_values2).tolist()

# The position of the bars on the x-axis
r = np.arange(len(y_values))

# Bar width
barWidth = 1
 
# Create brown bars
plt.bar(r, 
        y_values, 
        color='#7f6d5f', 
        edgecolor='white', 
        width=barWidth, 
        label=y_values_label)

# Create green bars (top), on top of the first ones
plt.bar(r, 
        y_values2, 
        bottom=y_values, 
        color='#557f2d', 
        edgecolor='white', 
        width=barWidth, 
        label=y_values2_label)

# Custom X axis
plt.xticks(r, x_values, fontweight='bold')
plt.xlabel("Year")

plt.legend()
plt.title(y_values_label + " and " + y_values2_label + " per capita GDP")


plt.show()
```
