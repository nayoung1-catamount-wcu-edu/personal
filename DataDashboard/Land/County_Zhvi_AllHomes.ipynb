```python
#Imports
import pandas as pd
```

```python
#Load Land data
df_zhvi = pd.read_excel('./Data/County_Zhvi_AllHomes.xlsx')

#Display table to ensure data loaded correctly
df_zhvi.head()
```

```python
#Filter data to NC
filter1 = df_zhvi['State'] == "NC"
df_zhvi_nc = df_zhvi[filter1]

#Check to ensure filter worked
df_zhvi_nc.head(5)
```

```python
#Change MunicipalCodeFIPS dtype to add leading 0's
df_zhvi_nc['MunicipalCodeFIPS'] = df_zhvi_nc['MunicipalCodeFIPS'].astype(str)
```

```python
#Add leading 0's and check to ensure they were added
df_zhvi_nc['MunicipalCodeFIPS'] = df_zhvi_nc['MunicipalCodeFIPS'].str.zfill(3)
df_zhvi_nc.head(5)
```

```python
#Save to csv file for export in Excel
df_zhvi_nc.to_csv('./Data/STG_ZLLW_County_Zhvi_AllHomes.txt')
```
