```python
import geopandas as gp
import matplotlib.pyplot as plt
```

```python
npr = "../Data/energy-pop-exposure-nuclear-plants-locations_reactors.shp"
```

```python
world = gp.read_file(gp.datasets.get_path('naturalearth_lowres'))
```

```python
reactors = gp.read_file(npr)
reactors.columns
```

```python
boundaries = world.boundary.plot(figsize=(12, 9))
reactors.plot(column='Status', ax=boundaries, markersize=10, legend=True)
```

```python
reactors.Country
```

```python
df_usa = reactors["Country"] == "UNITED STATES OF AMERICA"
df_usa = reactors[df_usa]
df_usa
```

```python
boundaries = world.boundary.plot(figsize=(12, 9))
df_usa.plot(ax=boundaries)
```
