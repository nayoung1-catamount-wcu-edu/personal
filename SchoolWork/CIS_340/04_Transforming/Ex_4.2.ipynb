#### capitals

```python
# Hardcode a dataframe named df_capitals
df_capitals = pd.DataFrame(
    [
    ["Ngerulmud",391,1.87],
    ["Vatican City",826,100],
    ["Yaren",1100,10.91],
    ["Funafuti",4492,45.48],
    ["City of San Marino",4493]
    ], 
    index = ["Palau", "Vatican City", "Nauru", "Tuvalu", "San Marino"],
    columns=['Capital', 'Population', 'Percentage'])

df_capitals.head()
```

```python
# Use loc to display the row for Tuvalu
df_capitals.loc['Tuvalu']
```

```python
# Use loc to display the rows for Nauru to San Marino
df_capitals.loc['Nauru':'San Marino']
```

```python
df_capitals.head()
```

```python
# Use iloc to display the row for Tuvalu
df_capitals.iloc[3]
```

```python
# Use iloc to display the rows for Nauru to San Marino
df_capitals.iloc[2:5]
```

```python
# Select rows where Percentage > 25
row_filter = df_capitals['Percentage'] > 25
# Row filter
df_capitals[row_filter].head()
```

#### grades

```python
df_grades = pd.DataFrame([[6, 4], [7, 8], [6, 7], [6, 5], [5, 2]], 
                       index = ['Mary', 'John', 'Ann', 'Pete', 'Laura'],
                       columns = ['test_1', 'test_2'])
df_grades
```

# Part 1:  IMDB database
Using data from: https://raw.githubusercontent.com/justmarkham/pandas-videos/master/data/imdb_1000.csv

```python
#1.Read the csv file into a pandas dataframe named df_movies
df_movies = pd.read_csv('https://raw.githubusercontent.com/justmarkham/pandas-videos/master/data/imdb_1000.csv')
#2.Display the first five records/rows in the dataframe
df_movies.head()
```

```python
# Using loc: Select the top 10 movies - based on star_rating 
df_movies.loc[:9]
```

```python
# Using iloc: Select the top 10 movies - based on star_rating 
df_movies.iloc[:9]
```

```python
# Select all the movies with an R rating
row_filter = df_movies['content_rating'] == 'R'

df_movies[row_filter].head()
```

```python
# Select all the movies with that are longer than three and a half hours
row_filter = df_movies['duration'] > 210

df_movies[row_filter].sort_values('duration', ascending = False)
```

```python
# All the Godfather movies??
godfather_index = df_movies['title'].str.contains('Godfather')

df_movies[godfather_index]
```

```python
# All the Charlton Heston movies??
heston_index = df_movies['actors_list'].str.contains('Heston')

df_movies[heston_index]
```

```python
# All the Jaws movies are in the top 250 listing?
search_index = df_movies['title'].str.contains('Jaws')

df_movies[search_index].head(0)
```

```python
df_movies.head()
```

```python
# All the movies Sean Connery has been in (top 250) 
search_index = df_movies['actors_list'].str.contains('Sean Connery')

df_movies[search_index].head(1)
```
