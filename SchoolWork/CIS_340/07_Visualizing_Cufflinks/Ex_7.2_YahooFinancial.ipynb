# In Class Chart 1  

- **Retrieve and plot the *Closing* prices for the following stocks**:  
  - Facebook  
  - Google  
  - Amazon  
  - Twitter
  
  


- **Start and End Dates**:  
  - Start:  Jan 1, 2014  
  - End:    Dec 31, 2018

### Set Start and End dates for Stocks to Retrieve  

```python
# Start for Stock series:  Jan 1, 2014
start = dt.datetime(2014, 1, 1)

# End of Stock series:  Dec 31, 2018
end = dt.datetime(2018, 12, 31)
```

### Retrieve the stocks

```python
stocks_to_retrieve = ['FB', 'GOOG', 'AMZN', 'TWTR']
df = web.DataReader(stocks_to_retrieve, 'yahoo', start, end)
df.head()
```

### Extract Closing Prices from Retrieved Stock Data

```python
df_closing = df['Close']
df_closing.head()
```

# Plot Data

### plot with pandas plot()

```python
df_closing.plot()
```

### Plot with cufflinks

```python
df_closing.iplot(kind='line',
                 title='7.2 Chart 1: Amazon, Facebook, Google and Twitter - After 2013')
```

# In Class Chart 2 
- Use your df_closing dataframe  
- Plot Google and Amazon 
- Use Theme:  space
- Use a Fill Chart 
- The X axis label should be:  Dates

# Ex Chart 1  
- Plot Closing prices for Google vs Amazon for 2018
- Chart Type: Spread  
  - Do this so the spread color is green rather than red

```python
start = dt.datetime(2017, 12, 31)
end = dt.datetime(2018, 12, 31)
index = ['GOOG', 'AMZN']
df = web.DataReader(index, 'yahoo', start, end)
df_closing = df['Close']
df_closing.head()

df_closing[index].iplot(kind='spread',
                    theme='space',
                     title='7.2 Ex Chart 1: Google and Amazon - 2018')
```

# Ex Chart 2  
- Plot **Highs** for Google and Amazon from 2014-2018 (you can start with the **df** dataframe)
  - **Note:  We are peeling out the *High* columns rather than the *Closing* columns**
- Chart Type: Line Chart with Bestfit Lines for both stocks


```python
start = dt.datetime(2014, 1, 1)
end = dt.datetime(2018, 12, 31)
index = ['GOOG', 'AMZN']
df = web.DataReader(index, 'yahoo', start, end)
df_high = df['High']
df_high.head()

df_high[index].iplot(kind='line',
                      bestfit=True,
                    theme='space',
                     title='7.2 Ex Chart 2: Google and Amazon (2014-2018)')
```

# Ex Chart 3  
- Plot Volume for Facebook and Twitter for 2017 and 2018
- Chart Type: Spread Chart with Bestfit line

```python
start = dt.datetime(2017, 1, 1)
end = dt.datetime(2018, 12, 31)
index = ['FB', 'TWTR']
df = web.DataReader(index, 'yahoo', start, end)
df_volume = df['Volume']
df_volume.head()

df_volume[index].iplot(kind='line',
                    theme='space',
                       bestfit=True,
                     title='7.2 Ex Chart 3: Volume for Facebook and Twitter (2017-2018)')
```

# Ex Chart 4  
- Fill a dataframe (df_open_close) with the difference in opening and closing prices for Google and Facebook
  - Note:  You can get that by creating a new dataframe that is df_closing - df_open 
- Do this just for the first quarter of 2018  
- Chart Type: Filled Line Chart

```python
start = dt.datetime(2018, 1, 1)
end = dt.datetime(2018, 3, 31)
index = ['GOOG', 'FB']
df = web.DataReader(index, 'yahoo', start, end)
df_open = df['Open']
df_open_close = (df['Close']-df['Open'])
df_open_close.head()

df_open_close[index].iplot(kind='line',
                        fill=True,
                    theme='space',
                     title='7.2 Ex Chart 4: Google and Facebook, Close-Open, First Quarter 2018')
```
